```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
#Load Dataset
iris = pd.read_csv('Iris.csv')
```

```python
iris.head()
```

```python
#Select the feature you want to visualize
feature_to_plot ="SepalLengthCm"
```

```python
#Create a Histogram of SepaLengthCm
plt.figure(figsize=(8,6))
sns.histplot(iris[feature_to_plot], kde=True, color="Blue")
plt.title(f"Histogram of {feature_to_plot} in Iris Dataset")
plt.xlabel(feature_to_plot)
plt.ylabel("Frequency")
plt.show()
```
